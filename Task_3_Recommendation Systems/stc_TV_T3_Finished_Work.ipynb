```python
df=dataframe.copy()
# Use get_dummies() to create one-hot encoding for movie names
dummies = pd.get_dummies(df['program_name'])
```

```python
dummies
```

```python
# Create a DataFrame for movie IDs
movie_ids = pd.DataFrame(dummies.columns, columns=['program_name'])
movie_ids['movie_id'] = range(1, len(movie_ids) + 1)

# Merge movie IDs back into the original DataFrame
df_with_ids = pd.merge(df, movie_ids, on='program_name')
```

```python
df_with_ids.tail(10)
```

```python
import re
#Remove any unneccessry characters from the program name
df_with_clean_titles =df_with_ids.copy()
def clean_title(program_name):
    program_name = re.sub("[^a-zA-Z0-9 ]", "", program_name)
    return program_name
```

```python
#apply the clean_title function
df_with_clean_titles["clean_program_name"] = df_with_clean_titles["program_name"].apply(clean_title)
```

```python
df_with_clean_titles
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
vectorizer = TfidfVectorizer(ngram_range=(1,2))

tfidf = vectorizer.fit_transform(df_with_clean_titles["clean_program_name"])
```

```python
from sklearn.metrics.pairwise import cosine_similarity
import numpy as np
#Clac the similarity between the searched title and the program_names
def search(title):
    title = clean_title(title)
    query_vec = vectorizer.transform([title])
    similarity = cosine_similarity(query_vec, tfidf).flatten()
    indices = np.argpartition(similarity, -5)[-5:]#return the 5 movies with greatest similarity
    results = df_with_clean_titles.iloc[indices].iloc[::-1]# reverse the order

    return results
```

```python
def find_similar_movies(movie_id):
  similar_users = df_with_clean_titles[(df_with_clean_titles["movie_id"] == movie_id) & (df_with_clean_titles["rating"] >= 4)]["user_id_maped"].unique()
  similar_user_recs = df_with_clean_titles[(df_with_clean_titles["user_id_maped"].isin(similar_users)) & (df_with_clean_titles["rating"] >= 4)]["movie_id"]

  similar_user_recs_count = similar_user_recs.value_counts() / len(similar_users)
  similar_user_recs_most_liked=similar_user_recs_count[similar_user_recs_count > .1]

  all_users = df_with_clean_titles[(df_with_clean_titles["movie_id"].isin(similar_user_recs_most_liked.index)) & (df_with_clean_titles["rating"] >= 4)]
  all_user_recs = all_users["movie_id"].value_counts() / len(all_users["user_id_maped"].unique())

  rec_percentages = pd.concat([similar_user_recs_most_liked, all_user_recs], axis=1)
  rec_percentages.columns = ["similar", "all"]

  rec_percentages["score"] = rec_percentages["similar"] / rec_percentages["all"]
  rec_percentages = rec_percentages.sort_values("score", ascending=False)


  df_programs=df_with_clean_titles.copy()

  # Keep only the relevant columns
  df_filtered = df_programs[['program_name', 'movie_id', 'program_genre', 'clean_program_name']]

  # Drop duplicates to keep distinct movie information
  df_unique = df_filtered.drop_duplicates()

  # Display the resulting DataFrame
  df_unique.head()

  return rec_percentages.head(10).merge(df_unique, left_index=True, right_on="movie_id")[["score", "program_name", "program_genre"]]

```
