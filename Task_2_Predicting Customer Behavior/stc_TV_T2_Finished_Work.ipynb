```python
df=dataframe.copy()
```

```python
df.columns
```

```python
# Convert date to datetime
df['date_'] = pd.to_datetime(df['date_'])
```

```python
# Create temporal features
df['day'] = df['date_'].dt.day
df['day_of_week'] = df['date_'].dt.dayofweek
df['month'] = df['date_'].dt.month
df['year'] = df['date_'].dt.year
```

```python
df.tail(25)
```

```python
df.describe()
```

Prophete

```python
from prophet import Prophet
new_df=df.copy()
new_df=new_df.rename(columns={'date_':'ds','Total_watch_time_in_houres':'y'})[['ds','y']]
new_df.head()
```

```python
model = Prophet(seasonality_prior_scale=0.5)
model = model.fit(new_df)
future = model.make_future_dataframe(periods=61, freq='D')
forecast = model.predict(future)
```

```python
print(forecast[['ds', 'yhat', 'yhat_lower', 'yhat_upper']])
```

```python
# Python
model.plot(forecast)
plt.axhline(y=0, color='red')
plt.title('Default Prophet');
```
