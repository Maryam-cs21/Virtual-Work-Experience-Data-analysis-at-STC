```python
# Data Preprocessing on the input data
dataframe = dataframe.drop(columns=['Column1'])         # dropping the index column
dataframe['program_name'] = dataframe['program_name'].str.strip()  # trim spaces in movies names to avoid misspellings in input data
dataframe['date_'] = pd.to_datetime(dataframe['date_'])  # read date column as date data type
dataframe[['duration_seconds', 'season','episode','series_title','hd']] = dataframe[['duration_seconds', 'season','episode','series_title','hd']].apply(pd.to_numeric)  # read numeric columns as numeric data types
dataframe[['user_id_maped', 'program_name','program_class','program_desc','program_genre','original_name']] = dataframe[['user_id_maped', 'program_name','program_class','program_desc','program_genre','original_name']].astype(str) # read string columns as string data types
```

```python
# make a copy of the dataframe for working on task 1
df=dataframe.copy()
```

```python
# Here we try to get the most watched movies (Total Views / Total Users Views / Total watch time)
# For series we concatenated the Session episode to differentiate between episodes
grouped1=df.copy()
grouped1.loc[grouped1['program_class'] == 'SERIES/EPISODES', 'program_name'] = grouped1['program_name']+'_SE'+grouped1['season'].astype(str)+'_EP'+grouped1['episode'].astype(str)
grouped1 = grouped1.groupby(['program_name','program_class'])\
.agg({'user_id_maped': [('co1', 'nunique'),('co2', 'count')],\
      'duration_seconds': [('co3', 'sum')] }).reset_index()
grouped1.columns = ['program_name','program_class','No of Users who Watched', 'No of watches', 'Total watch time in seconds']
grouped1['Total watch time in houres']=grouped1['Total watch time in seconds']/3600
grouped1 = grouped1.drop(columns=['Total watch time in seconds'])
grouped1 = grouped1.sort_values(by=['Total watch time in houres', 'No of watches','No of Users who Watched'], ascending=False).reset_index(drop=True)
```

```python
# Here we try to study the customer experience against Program class
grouped2=df.copy()
grouped2 = grouped2.groupby('program_class')\
.agg({'user_id_maped': [('co1', 'nunique'),('co2', 'count')],\
      'duration_seconds': [('co3', 'sum')] }).reset_index()
grouped2.columns = ['program_class','No of Users who Watched', 'No of watches', 'Total watch time in seconds']
grouped2['Total watch time in houres']=grouped2['Total watch time in seconds']/3600
grouped2 = grouped2.drop(columns=['Total watch time in seconds'])
grouped2 = grouped2.sort_values(by=['Total watch time in houres', 'No of watches','No of Users who Watched'], ascending=False).reset_index(drop=True)
```

```python
"""
TODO study the relation and user's behaviour Against HD flag, feel free to include any comparison
"""
# Count how many times each program was watched with HD flag
grouped3=df.copy()
grouped3.loc[grouped3['program_class'] == 'SERIES/EPISODES', 'program_name'] = grouped3['program_name']+'_SE'+grouped3['season'].astype(str)+'_EP'+grouped3['episode'].astype(str)

grouped3 = grouped3.groupby(['program_name', 'program_class'])\
    .agg({'user_id_maped': [('Total Views', 'count'), ('No of times watched with HD', lambda x: (x[grouped3['hd'] == 1]).count()),
     ('No of times watched without HD', lambda x: (x[grouped3['hd'] == 0]).count())]}).reset_index()

grouped3.columns = ['program_name', 'program_class', 'Total Views', 'No of times watched with HD','No of times watched without HD']

grouped3 = grouped3.sort_values(by=['No of times watched with HD'], ascending=False).reset_index(drop=True)
```

```python
grouped3.head(10)
```

```python
import plotly.express as px
import pandas as pd

# Sample data from the grouped DataFrame
data = {
    'program_name': grouped3['program_name'].head(10),
    'Total Views': grouped3['Total Views'].head(10),
    'No of times watched with HD': grouped3['No of times watched with HD'].head(10),
    'No of times watched without HD': grouped3['No of times watched without HD'].head(10)
}

# Creating a DataFrame
df2 = pd.DataFrame(data)

# Melt the DataFrame to long format for Plotly
df_melted = df2.melt(id_vars='program_name',
                    value_vars=['Total Views', 'No of times watched with HD', 'No of times watched without HD'],
                    var_name='View Type', value_name='Count')

# Create the bar chart
fig = px.bar(df_melted,
             x='program_name',
             y='Count',
             color='View Type',
             barmode='group',
             title='Total Users Watching by Program',
             labels={'Count': 'Number of Views', 'program_name': 'Program Name'})

# Show the plot
fig.show()
```

```python
# Count how many times each program was watched with HD flag
grouped4=df.copy()

grouped4 = grouped4.groupby(['program_genre'])\
    .agg({
        'user_id_maped': [('Total Views', 'count'),
                          ('No of times watched with HD', lambda x: (x[grouped4['hd'] == 1]).count()),
                          ('No of times watched without HD', lambda x: (x[grouped4['hd'] == 0]).count())],
        'duration_seconds': [('Total Time Watched with HD', lambda x: x[grouped4['hd'] == 1].sum()),
                               ('Total Time Watched without HD', lambda x: x[grouped4['hd'] == 0].sum())]
    }).reset_index()

grouped4.columns = ['program_genre','Total Views',
                   'No of times watched with HD', 'No of times watched without HD',
                   'Total Time Watched with HD (Seconds)', 'Total Time Watched without HD (Seconds)']

grouped4['Total Time Watched with HD (Hours)'] = grouped4['Total Time Watched with HD (Seconds)'] / 3600
grouped4['Total Time Watched without HD (Hours)'] = grouped4['Total Time Watched without HD (Seconds)'] / 3600

grouped4=grouped4.drop (columns=['Total Time Watched with HD (Seconds)'])
grouped4=grouped4.drop (columns=['Total Time Watched without HD (Seconds)'])

grouped4 = grouped4.sort_values(by=['No of times watched with HD'], ascending=False).reset_index(drop=True)

grouped4.head(20)
```

```python
import plotly.express as px
import pandas as pd

# Assuming 'grouped' DataFrame is already created as per your provided code

# Create the histogram
df_melted2 = grouped4.melt(id_vars='program_genre',
                          value_vars=['Total Views', 'No of times watched with HD', 'No of times watched without HD'],
                          var_name='View Type', value_name='Count')

# Create the bar chart
fig = px.bar(df_melted2,
             x='program_genre',
             y='Count',
             color='View Type',
             barmode='group',
             title='Total Views by Program Genre',
             labels={'Count': 'Number of Views', 'program_genre': 'Program Genre'})

# Show the plot
fig.show()
```
